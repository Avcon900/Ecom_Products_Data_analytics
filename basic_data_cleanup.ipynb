```python
import pandas as pd

df = pd.read_csv('data/ecommerce_scraped_data_final.csv')

df.head()
```

```python
# Fix common encoding issues
df['Title'] = df['Title'].str.encode('utf-8').str.decode('utf-8', 'ignore')
df['MRP'] = df['MRP'].str.replace('â‚¹', '₹', regex=False)
df['Price After Discount'] = df['Price After Discount'].str.replace('â‚¹', '₹', regex=False)

df.head()
```

```python
# Convert 'Rating' to numeric, extracting the numeric part
df['Rating'] = df['Rating'].str.extract(r'([\d.]+)').astype(float)

df.head()
```

```python
# Convert 'Number of Reviews' to numeric, removing commas and converting to int

# Remove commas
df['Number of Reviews'] = df['Number of Reviews'].str.replace(',', '', regex=False)

# Convert to float first, then fill missing with 0, then convert to int
df['Number of Reviews'] = df['Number of Reviews'].astype(float).fillna(0).astype(int)

df.head()
```

```python
# Convert 'Price After Discount' and 'MRP' to numeric, removing currency symbols and commas

df['Price After Discount'] = df['Price After Discount'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['MRP'] = df['MRP'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float)

df.head()
```

```python
import ast

# Step 1: Safely convert stringified dict to Python dict
def safe_eval(x):
    try:
        return ast.literal_eval(x) if pd.notnull(x) else {}
    except:
        return {}

df['Star Rating Percentage'] = df['Star Rating Percentage'].apply(safe_eval)

# Step 2: Expand dictionary into columns
star_df = df['Star Rating Percentage'].apply(pd.Series)

# Step 3: Rename columns
star_df = star_df.rename(columns={
    5: '5 Star Percentage',
    4: '4 Star Percentage',
    3: '3 Star Percentage',
    2: '2 Star Percentage',
    1: '1 Star Percentage'
})

# Step 4: Concatenate back to original DataFrame (optional)
df = pd.concat([df.drop(columns=['Star Rating Percentage']), star_df], axis=1)
```

```python
import re

def parse_bought_last_month(value):
    if pd.isnull(value):
        return 0
    value = value.strip().lower()
    
    match = re.match(r'(\d+(?:\.\d+)?)(k)?\+?', value)
    
    if match:
        number = float(match.group(1))
        if match.group(2) == 'k':
            return int(number * 1000)
        else:
            return int(number)
    else:
        return 0  # or np.nan if you prefer

df['Bought Last Month (Numeric)'] = df['Bought Last Month'].apply(parse_bought_last_month)
# Drop the original 'Bought last Month' column if no longer needed

df.head()
```

```python
df.to_csv("data/controllers_cleaned.csv", index=False)
```
